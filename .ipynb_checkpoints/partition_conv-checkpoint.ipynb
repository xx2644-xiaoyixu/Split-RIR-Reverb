```python
def partition_ir(h, block_size):
    """
    Split impulse response into uniform partitions of length block_size,
    and precompute their FFTs with FFT size = 2 * block_size.

    Parameters
    ----------
    h : (np.ndarray)
        Impulse response, 1D array.
        
    block_size : (int)
                The block or buffer size, usually equals to 2**n
        
    Returns
    -------
    H_parts : List[np.ndarray]
    A list containing the FFT of each partitioned block of the impulse response.
    Each element corresponds to one partition in the frequency domain.
    """
    fft_size = 2 * block_size # if block size of input sig= block size of h_part,output full mode conv sig len is 2B-1

    num_parts = int(np.ceil(len(h) / block_size))
    H_parts = []

    for i in range(num_parts):
        start = i * block_size
        end = min(start + block_size, len(h))
        
        h_part = h[start:end]
        pad_len = fft_size - len(h_part)
        h_part_p = np.concatenate([h_part, np.zeros(pad_len)])

        H_parts.append(np.fft.rfft(h_part_p))

    return H_parts


def partitioned_conv(x, h, block_size):
    """
    Uniform partitioned convolution.

    Parameters
    ----------
    x : np.ndarray
        Input signal, 1D.
    h : np.ndarray
        Impulse response, 1D.
    block_size : int
        Uniform partition size / processing block size.

    Returns
    -------
    y : np.ndarray
        Full convolution result, length len(x) + len(h) - 1.
    """
    if x.ndim != 1 or h.ndim != 1:
        raise ValueError("x and h must be 1D arrays.")

    N = len(x)
    K = len(h)
    fft_size = 2 * block_size

    # Precompute FFT of each IR partition
    H_parts = partition_ir(h, block_size)
    num_parts = len(H_parts)

    # Number of input blocks
    num_blocks = int(np.ceil(N / block_size))

    # Pad input so each block has exactly block_size samples
    x_padded = np.pad(x, (0, num_blocks * block_size - N))

    # Store recent input spectra (one per block)
    X_hist = [np.zeros(fft_size // 2 + 1, dtype=complex) for _ in range(num_parts)]

    # Output buffer for block accumulation
    y = np.zeros(num_blocks * block_size + K - 1)

    # Overlap from previous IFFT
    overlap = np.zeros(block_size)

    for b in range(num_blocks):
        start_time = time.time()
        block_start = b * block_size
        x_block = x_padded[block_start:block_start + block_size]

        # Zero-pad current input block to FFT size
        x_fft_in = np.zeros(fft_size)
        x_fft_in[:block_size] = x_block
        X_now = np.fft.rfft(x_fft_in)

        # Update input spectrum history
        X_hist = [X_now] + X_hist[:-1]

        # Frequency-domain accumulation
        Y_freq = np.zeros_like(X_now)
        for p in range(num_parts):
            Y_freq += X_hist[p] * H_parts[p]

        # Back to time domain
        y_time = np.fft.irfft(Y_freq, n=fft_size)

        # Overlap-add style output
        y_block = y_time[:block_size] + overlap
        overlap = y_time[block_size:]

        y[block_start:block_start + block_size] += y_block
        end_time = time.time()
        print(f"partition conv took {end_time - start_time} seconds.")

    # Add final overlap
    tail_start = num_blocks * block_size
    y[tail_start:tail_start + block_size] += overlap

    # Trim to exact full convolution length
    y = y[:N + K - 1]
    return y
```
